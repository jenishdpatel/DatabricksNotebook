```
import datetime

x = datetime.datetime(2022, 3, 29)
print(x.strftime("%A"))  #full name of week name
print(x.strftime("%B"))  #full name of month name
print(x.strftime("%d"))  #get date
print(x.strftime("%H"))  #get hour
print(x.strftime("%m"))  #get month number
print(x.strftime("%M"))  #get minute
print(x.strftime("%p"))  #get am/pm
print(x.strftime("%S"))  #get second
print(x.strftime("%u"))  #week day number
print(x.strftime("%V"))  #total week in this year
print(x.strftime("%x"))  #get date
print(x.strftime("%X"))  #get time
print(x.strftime("%Y"))  #get year
```
